# Bank ATM Project

```python
import sys

class Atm:
    customer_pin = 2356

    def __init__(self, name):
        self.bank_name = name
        self.balance = 250

    def start(self):
        print(
            f"\n******************** Welcome to {self.bank_name} Bank! ********************")
        self.enter_pin(self.customer_pin)

    def enter_pin(self, pin):
        self.pin = pin
        count = 0
        while count < 3:
            self.pin = int(input("\nPlease enter your 4 digit PIN: "))
            if self.pin == self.customer_pin:
                self.menu()
            else:
                print("Incorrect PIN")
                count += 1
                if count == 3:
                    print("\nYour card has been blocked!")
                    print("Please contact our Customer Care to unblock your card.\n")

    def menu(self):
        print("""
Welcome! What would you like to do today?

******************** ATM MENU ********************
1. Check Balance
2. Deposit
3. Withdraw
4. Exit
**************************************************
""")

        while True:
            selection = int(input('Enter 1, 2, 3 or 4 to make a selection: '))
            if selection == 1:
                self.check_balance()
            elif selection == 2:
                while True:
                    self.dep_amount = float(
                        input("\nEnter an amount you would like to deposit: "))
                    self.deposit(self.dep_amount)
            elif selection == 3:
                while True:
                    self.with_amount = float(
                        input("\nEnter an amount you would like to withdraw: "))
                    self.withdraw(self.with_amount)
            elif selection == 4:
                print("\nBye! See you another time...\n")
                sys.exit()
            else:
                print("Invalid selection\n")

    def check_balance(self):
        print(f"\nYour balance is: £{self.balance}")
        self.new_transaction()

    def deposit(self, amount):
        self.dep_amount = amount
        if self.dep_amount < 1:
            print("Please enter an amount greater than or up to £1")
        else:
            self.balance += self.dep_amount
            print("\nDeposit successful!")
            print(f"Your balance is £{self.balance}")
            self.new_transaction()

    def withdraw(self, with_amount):
        self.with_amount = with_amount
        if self.balance == 0:
            print(f"Your balance is {self.balance}")
            print("You have to deposit some money before you can make a withdrawal.")
            self.new_transaction()
        elif self.with_amount >= 0 and self.with_amount < 5:
            print("Please enter an amount greater than or up to £5")
        elif self.with_amount > self.balance:
            print("\nInsufficient Funds!")
            print(f"Your balance is: £{self.balance}")
            print(f"Please enter an amount less than or up to £{self.balance}")
        elif self.balance == 0:
            print("You have to deposit some money before you can make a withdrawal.")
            self.new_transaction()
        elif self.with_amount <= self.balance:
            self.balance -= self.with_amount
            print("Withdrawal successful!")
            print(f"Your balance is £{self.balance}")
            self.new_transaction()

    def new_transaction(self):
        while True:
            response = input(
                "\nWould you like to perform another transaction (y/n): ")
            if response.lower() == 'y':
                self.menu()
            elif response.lower() == 'n':
                print("\nBye! See you another time...\n")
                sys.exit()
            else:
                print("Invalid selection. Please enter 'y' or 'n'\n")
```

```python
if __name__ == '__main__':
    atm = Atm("HSBC")
    atm.start()
```
